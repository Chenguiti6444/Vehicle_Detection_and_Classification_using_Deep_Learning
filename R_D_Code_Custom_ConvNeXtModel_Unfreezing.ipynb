```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from google.colab import drive

drive.mount('/content/drive')

# Load the saved model
saved_model_path = r'/content/drive/MyDrive/R&D Project/Models/Customized_ConvNeXt_Model.tf'
model = tf.keras.models.load_model(saved_model_path)
```

```python
# Unfreezing some layers in the model
for layer in model.layers:
    if 'batch_normalization' not in layer.name.lower():
        layer.trainable = True
    else:
        layer.trainable = False

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Save the recompiled, unfrozen model under a new name
new_saved_model_path = r'/content/drive/MyDrive/R&D Project/Models/Customized_ConvNeXt_Model_Unfrozen.tf'
model.save(new_saved_model_path)
```
